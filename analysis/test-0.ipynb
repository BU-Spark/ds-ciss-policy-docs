```python
import pandas as pd
df = pd.read_csv('./cate.csv')
df.head()
```

```python
categories = df['type'].unique()
print(df.shape, len(categories))
```

```python
from collections import Counter
import matplotlib.pyplot as plt
c = Counter(df['type'])
freq = c.most_common()
freq[:20]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 8))
plt.bar(*zip(*freq[1:]))
plt.show()
```
