```python
import os
cwd = os.path.abspath('')
```

```python
fileCount = 0
docUnique = {}
regions = next(os.walk(cwd))[1]
for region in regions:
    years = next(os.walk(os.path.join(cwd, region)))[1]
    for year in years:
        docs = next(os.walk(os.path.join(cwd, region, year)))[2]
        for doc in docs:
            if doc in docUnique:
                print(doc, docUnique[doc], '{}/{}'.format(region, year))
            else:
                docUnique[doc] = '{}/{}'.format(region, year)
```
