```python
import os
```

```python
fileCount = 0
file = dict()
for r, d, files in os.walk("./"):
    for f in files:
        if f.endswith('.txt'):
            # priint file name
            if f not in file:
                file[f] = 1
            else:
                file[f] += 1
            fileCount += 1
print(f"Total number of .txt files: {fileCount}, and the unique files name count is: {len(file)}")
```

```python
# show all value >1
for k, v in file.items():
    if v > 1:
        print(f"File name: {k} has {v} counts")
```
